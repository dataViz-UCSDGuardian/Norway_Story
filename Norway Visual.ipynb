```python
years = []
for i in range (2003,2015):
    years.append(i)
    
print(years)
```

```python
welfare = [6186,6265,6456,6643,7043,7115,7173,7418,7750,8288,8414,8892]
careOrder = [1135,1175,1435,1486,1726,1653,1665]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,9))
graph = plt.subplot(111)
graph.get_xaxis().tick_bottom()
graph.get_yaxis().tick_left()

plt.plot(years,welfare, color = 'lightsalmon')
plt.plot(years[5:len(years)],careOrder, color = 'violet')

plt.xlabel('Years')
plt.ylabel('Number of Children under All Services')
plt.title('Children under Welfare Services in Norway')

```
